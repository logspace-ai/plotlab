PlotLab is a plotting library for Python created on top of Plotly. It is designed to be as simple as it can get, reducing boilerplate code in daily tasks which often tend to be tedious.

### QuickStart

```python
import numpy as np
import pandas as pd
from plotlab import Figure
```

```python
# Start with any dataframe
df = pd.DataFrame({'x': np.random.randn(100), 'y': np.random.randn(100)},   index=list(range(100)))
```

```python
# Create a figure and stack plots.
fig = Figure()
fig.bar(df.x)
fig.line(df.y)
fig.show()
```

```python
# Quickly add annotations
fig.annotation(14, 1.7, 'Hello World')
fig.show()
```

```python
# Reset the figure...
fig.clear()

# and start over.
fig.area(df.x)
fig.show()
```

```python
fig.scatter(df.x)
fig.show()
```

```python
# Add threshold lines
fig.threshline(df.x, 2, -2, 'lime', 'red')
fig.show()
```

```python
# Paint the background
fig.background(df.x.iloc[:10], 'Hello', 'red', 0.5)
fig.background(df.x.iloc[70:80], 'World', 'lime', 0.5)
fig.show()
```
