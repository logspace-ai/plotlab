```python
import sys
sys.path.append('../')

import numpy as np
import pandas as pd
from plotlab import Figure
```

```python
df = pd.DataFrame({'x': np.random.randn(100), 'y': np.random.randn(100)},   index=list(range(100)))
```

```python
fig = Figure()
fig.bar(df.x)
fig.line(df.y)
fig.show()
```

```python
fig.annotation(14, 1.7, 'Hello World')
fig()
```

```python
fig.clear()
fig.area(df.x)
fig()
```

```python
fig.scatter(df.x)
fig()
```

```python
fig.clear()
fig.dotline(df.x)
fig()
```

```python
fig.clear()
fig.background(df.x, 'Hello', 'red', 0.5)
fig()
```

```python
fig.hline(2)
fig.hline(3)
fig()
```

```python
fig.clear()
fig.line(df.x)
fig.threshline(df.x, 1, -1, 'lime', 'red')
fig()
```
